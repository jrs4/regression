# Regression Model - Wheat Calibration

Artificial Neural Network based on a TensorFlow / Keras model

**Training Data pre-treatment**:
- Load calibration data for training
- Only use pixels from 9 to 33 (25 pixels) for training
- Drop any rows with y values outside limits (5-20% for protein in wheat)
- Fit the training data into Principal Component Analysis (PCA) from scikit-learn
- Transform the training data into principal components
- Use MinMaxScaler() to fit the data between (-1, 1)

**Validation Data pre-treatment**:
- Load validation data
- Only use pixels from 9 to 33 (25 pixels) for training
- Drop any rows with y values outside limits (5-20% for protein in wheat)
- Transform the validation data into principal components using the PCA model from the training set
- Use MinMaxScaler() to fit the data between (-1, 1)

**Train Model in TensorFlow / Keras**:

After the data is ready for the neural network, create the model in Keras, compile it and fit it.

**Testing Data pre-treatment**:
- Same as validation data pre-treatment

**Check predictions and calculate slope and biases**

The training and validation data usually come from the master system whereas the testing data is usually the prediction made on other systems to check calibration transferability

### Import Libraries
`tensorflow`, `numpy`, `pandas`, `matplotlib`, `sklearn`

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

### Training Data

```python
#Training Data
path = "./Data/(1) Training/"
all_files = glob.glob(os.path.join(path, "*.csv")) #make list of paths

li = []

for filename in all_files:
    dfn = pd.read_csv(filename, index_col=None, header=0)
    li.append(dfn)

df_train = pd.concat(li, axis=0, ignore_index=True, sort=False)

#rename "cal1" "cal2" to "Protein" "Moisture"
df_train = df_train.rename(index=str, columns={"Cal1": "Protein", "Cal2": "Moisture"})
df_train = df_train.rename(index=str, columns={"cal1": "Protein", "cal2": "Moisture"})

#randomise rows
df_train = df_train.sample(frac=1).reset_index(drop=True)

#Drop values of spectra if NaN
for i in range(9,33):
    df_train = df_train.dropna(subset=["Sp{}".format(i)])

#Drop values of protein if NaN or outside ranges
df_train = df_train.dropna(subset=["Protein"])
df_train = df_train[df_train.Protein < 20]
df_train = df_train[df_train.Protein > 5]
if "Exclude" in df_train:
    df_train = df_train[df_train.Exclude != 'Y']
    
train_spectra = df_train.loc[:,"Sp9":"Sp33"].values
train_spectra_norm = train_spectra.copy()

#for value, spectra in enumerate(train_spectra, 0):
#    train_spectra_norm[value] = (spectra - spectra.mean()) / (spectra.std())
#    
#train_spectra_norm = (train_spectra - train_spectra.mean()) / (train_spectra.max() - train_spectra.min())

train_values = df_train["Protein"].values
df_train.head()
```

```python
print(train_spectra.shape)
print(train_values.shape)
print(np.argwhere(np.isnan(train_values)))

plt.plot(train_spectra.T)
plt.title('Spectra')
plt.ylabel('Abs')
plt.xlabel('Samples')
plt.show()

plt.plot(train_values)
plt.title('Protein')
plt.ylabel('%')
plt.xlabel('Wave')
plt.show()
```

#### Principal Component Analysis (PCA) on training set

#### Data scaled to (-1,1) with MinMaxScaler

```python
calibration_mean = train_spectra.mean(axis=0)
calibration_std = train_spectra.std(axis=0)

pca_components = 14

pca = PCA(n_components=pca_components)
pca.fit(train_spectra)
train_spectra_pca = pca.transform(train_spectra)

train_spectra_norm = train_spectra.copy()
train_spectra_norm = (train_spectra_norm - calibration_mean)/calibration_std

scaler = MinMaxScaler(feature_range=(-1,1))
train_spectra_pca = scaler.fit_transform(train_spectra_pca)
```

```python
plt.plot(train_spectra_pca.T)
plt.title('Train Spectra')
plt.ylabel('Abs')
plt.xlabel('Wave')
plt.show()
```

### Validation Data

```python
#Validation Data
path = "./Data/(2) Validation/"
all_files = glob.glob(os.path.join(path, "*.csv")) #make list of paths

li = []

for filename in all_files:
    dfn = pd.read_csv(filename, index_col=None, header=0)
    li.append(dfn)

df_val = pd.concat(li, axis=0, ignore_index=True, sort=False)

#rename "cal1" "cal2" to "Protein" "Moisture"
df_val = df_val.rename(index=str, columns={"Cal1": "Protein", "Cal2": "Moisture"})
df_val = df_val.rename(index=str, columns={"cal1": "Protein", "cal2": "Moisture"})

#randomise rows
df_val = df_val.sample(frac=1).reset_index(drop=True)

#Drop values of spectra if NaN
for i in range(9,33):
    df_val = df_val.dropna(subset=["Sp{}".format(i)])
    
#Drop values of protein if NaN
df_val = df_val.dropna(subset=["Protein"])
df_val = df_val[df_val.Protein < 20]
df_val = df_val[df_val.Protein > 5]
if "Exclude" in df_val:
    df_val = df_val[df_val.Exclude != 'Y']

val_spectra = df_val.loc[:,"Sp9":"Sp33"].values
val_spectra_norm = val_spectra.copy()

#for value, spectra in enumerate(val_spectra, 0):
#    val_spectra_norm[value] = (spectra - spectra.mean()) / (spectra.std())
    
#val_spectra_norm = (val_spectra - val_spectra.mean()) / (val_spectra.max() - val_spectra.min())

val_values = df_val["Protein"].values
```

```python
print(val_spectra.shape)
print(val_values.shape)
print(np.argwhere(np.isnan(val_values)))

plt.plot(val_spectra.T)
plt.title('Spectra')
plt.ylabel('Abs')
plt.xlabel('Samples')
plt.show()

plt.plot(val_values)
plt.title('Protein')
plt.ylabel('%')
plt.xlabel('Samples')
plt.show()
```

```python
val_spectra_pca = pca.transform(val_spectra)
print(val_spectra_pca.shape)

val_spectra_norm = val_spectra.copy()
val_spectra_norm = (val_spectra_norm - calibration_mean)/calibration_std

val_spectra_pca = scaler.transform(val_spectra_pca)
```

```python
plt.plot(val_spectra_pca.T[:])
plt.title('Validation Spectra')
plt.ylabel('Abs')
plt.xlabel('Wave')
plt.show()
```

### TensorFlow model
- Input layer = 14 nodes because we are using 14 principal components
- Hidden layer = 5 nodes with L2 regularization and an `tanh` activation function for slower learning
- Output layer = single node with ReLU activation function

```python
#model = tf.keras.models.Sequential([
#  tf.keras.layers.Dense(pca_components, input_dim = pca_components, kernel_initializer='normal', activation='relu'),
#  tf.keras.layers.Dense(50, kernel_initializer='normal', activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
#  tf.keras.layers.Dropout(0.8, noise_shape=None, seed=None),
#  tf.keras.layers.Dense(20, kernel_initializer='normal', activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
#  tf.keras.layers.Dense(20, kernel_initializer='normal', activation='tanh'),
#  tf.keras.layers.Dense(1, activation='relu')
#])


model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(pca_components, input_dim = pca_components, kernel_initializer='normal', activation='relu'),
  tf.keras.layers.Dense(5, kernel_initializer='normal', activation='tanh', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
  tf.keras.layers.Dense(1, activation='relu')
])

optimizer = tf.keras.optimizers.Adam(lr=0.0005, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae','acc'])
```

```python
%%time
hist = model.fit(train_spectra_pca, train_values, epochs=2000, verbose=1, validation_data=(val_spectra_pca,val_values))
```

```python
model.evaluate(val_spectra_pca, val_values)
```

```python
prediction = model.predict(val_spectra_pca)
```

In the first 100 epochs, the loss is very high, therefore the history plot starts from epoch 101

```python
plt.plot(hist.history['loss'][100:])
plt.plot(hist.history['val_loss'][100:])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```

```python
np.savetxt("./Model Prediction.csv", prediction, delimiter=',')
```

```python
#Testing Data
path = "./Data/(3) Testing/"
all_files = glob.glob(os.path.join(path, "*.csv")) #make list of paths

li = []

for filename in all_files:
    dfn = pd.read_csv(filename, index_col=None, header=0)
    li.append(dfn)

df_test = pd.concat(li, axis=0, ignore_index=True, sort=False)

#rename "cal1" "cal2" to "Protein" "Moisture"
df_test = df_test.rename(index=str, columns={"Cal1": "Protein", "Cal2": "Moisture"})
df_test = df_test.rename(index=str, columns={"cal1": "Protein", "cal2": "Moisture"})

#Drop values of spectra if NaN
for i in range(9,33):
    df_test = df_test.dropna(subset=["Sp{}".format(i)])
    
#Drop values of protein if NaN
df_test = df_test.dropna(subset=["Protein"])

test_spectra = df_test.loc[:,"Sp9":"Sp33"].values
test_values = df_test["Protein"].values

test_spectra_pca = pca.transform(test_spectra)

test_spectra_pca = scaler.transform(test_spectra_pca)

prediction = model.predict(test_spectra_pca)
```

```python
print("Lab Values = {}".format(test_values))
print("Prediction = {}".format(np.around(prediction.reshape(test_values.shape[0]),decimals=1)))
plt.scatter(test_values, prediction)
plt.title('Protein Prediction')
plt.ylabel('Predicted')
plt.xlabel('True')
plt.show()
```

```python
LR_X = prediction[:,0].reshape(prediction[:,0].shape[0],1)
LR_y = test_values.reshape(test_values.shape[0],1)
reg = LinearRegression().fit(LR_X, LR_y)
print ("R2 = " + str(reg.score(LR_X, LR_y)))

print("Slope = " + str(reg.coef_))
print("Bias = " + str(reg.intercept_))

predictionAdjusted = prediction[:,0] * reg.coef_ + reg.intercept_
print("Lab Values = {}".format(np.round(test_values,1)))
print("Prediction = {}".format(np.round(predictionAdjusted[0],1)))

error = (predictionAdjusted - test_values)
print("Error = {}".format(np.round(error,1)))

sep = error.std()
print("Standard Deviation = " + str(sep))
```

```python
plt.plot(error[0])
plt.title('Prediction Error')
plt.ylabel('Error %')
plt.show()
```
